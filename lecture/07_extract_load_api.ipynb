```python
import requests
import json
```

```python
url = 'https://jobsearch.api.jobtechdev.se'
url_for_search = f"{url}/search"

headers = {'accept': 'application/json'}
search_params = {'q': "analytics engineer", 'limit': 100}

response = requests.get(url_for_search, headers=headers, params=search_params)
response.raise_for_status()  # check for http errors
#all_job_ads = json.loads(response.content.decode('utf8'))
all_job_ads = response.json()["hits"]
#all_job_ads[0].keys()
for ads in all_job_ads:
    print(f"{ads["headline"]} - {ads["employer"]["name"]}")
```

### Generator function
- lazy expression
- memory efficient

```python
def nr_generator_function():
    for i in range(5):
        yield i
```

```python
#call for the number generator function to create a number generator
nr_generator = nr_generator_function()
```

```python
#lazy expression: cannot see contents before using it
nr_generator
```

```python
#use it as a list - first time
list(nr_generator)
```

```python
#use it as a list - second time
#we cannot use the content for the second time --> memory efficient
#this is for reducing temporary memory
list(nr_generator)
```

```python
#testing with a regular return, state not saved
def nr_not_generator_function():
    for i in range(5):
        return i
not_nr_generator = nr_not_generator_function()
not_nr_generator
```

### dlt is using yield in order to be memory effcientt. They require the use of yield

```python
#from JOBTECH API code examples, which can be simplified
def _get_ads(url_for_search, params):
    headers = {"accept": "application/json"}
    response = requests.get(url_for_search, headers=headers, params=params)
    response.raise_for_status()  # check for http errors
    return json.loads(response.content.decode("utf8"))
```

```python
#job genenarator example
def jobads_resource(params):

    url = "https://jobsearch.api.jobtechdev.se"
    url_for_search = f"{url}/search"

    for ad in _get_ads(url_for_search, params)["hits"]:
        yield ad
```

```python
job_generator = jobads_resource(search_params)
```

```python
#list out job for the first time
list(job_generator)
```
